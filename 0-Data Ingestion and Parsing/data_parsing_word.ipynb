```python
from langchain.document_loaders import UnstructuredWordDocumentLoader,Docx2txtLoader
```

### METHOD # 1 : Docx2txtLoader

```python
try:
        loader = Docx2txtLoader("/home/hammadali08/Downloads/0-DataIngestParsing/data/word_files/proposal.docx")
        data = loader.load()
        print(f"Number of documents: {len(data)}")
        print(f"First document content: {data[0].page_content[:500]}")  # Print first 500 characters of the first document
        print(f"First document metadata: {data[0].metadata}")
except Exception as e:
        print(f"An error occurred: {e}")
```

### Method # 2: Unstructured Word Loader

```python
try:
        loader = UnstructuredWordDocumentLoader("/home/hammadali08/Downloads/0-DataIngestParsing/data/word_files/proposal.docx",
                                                mode='elements')### Elements to preserved the structure of the document
        data = loader.load()
        print(f"Number of documents: {len(data)}")
        print(f"First document content: {data[0].page_content}")  # Print first 500 characters of the first document
        print(f"First document metadata: {data[0].metadata}")
except Exception as e:
        print(f"An error occurred: {e}")
```
